```python
rainfall=pd.read_csv(r'C:\Users\Tarun\Desktop\comp\mini-pred\MiniProjectDatasets\Rainfall_data_LL.csv')
#Coastal Karnataka,North Interior Karnataka - > Mandya ,South Interior Karnataka
```

```python
mandya=rainfall.loc[rainfall['SUBDIVISION'] == 'South Interior Karnataka']
```

```python
mandya.head()
```

```python
mandya.tail()
```

```python
mandya.columns
```

```python
mandya=mandya.drop(['Name','SUBDIVISION','ANNUAL','Jan-Feb','Mar-May', 'June-September', 'Oct-Dec', 'Latitude', 'Longitude'],axis=1)
```

```python
mandya.head()
```

```python
mandya['YEAR'] = pd.to_datetime(mandya['YEAR'], format='%Y')
df= mandya.reset_index(drop=True)
df.head(10)
```

```python
df.isnull().sum()
```

```python
from fbprophet import Prophet
```

```python
# January forecasting
jan = df[['JAN','YEAR']]
jan.columns=['y','ds']
jan.head(100)
```

```python
m = Prophet(interval_width=0.95)
model = m.fit(jan)
```

```python
future = m.make_future_dataframe(periods=10,freq='Y')
forecast = m.predict(future)
forecast.tail(10)
```

```python
forecast.loc[122]
```

```python
yhat_lower_jan=forecast.loc[122]['yhat_lower']
yhat_upper_jan=forecast.loc[122]['yhat_upper']
yhat_jan=forecast.loc[122]['yhat']
forecast_jan=[yhat_lower_jan,yhat_upper_jan,yhat_jan,'January']
print(forecast_jan)
```

```python
plot1 = m.plot(forecast)
```
