```python
df_mysore = df[df['Dist_Name'] == 'Mysore']
```

```python
from fbprophet import Prophet
```

```python
# May forecasting
may = df_mysore[['Water_per_hectare','Year']]
may.columns=['y','ds']
may.head(15)
```

```python
m = Prophet(interval_width=0.95)
model = m.fit(may)
future = m.make_future_dataframe(periods=3,freq='Y')
forecast = m.predict(future)
forecast.head(10)
```

```python
plot1 = m.plot(forecast)
```
